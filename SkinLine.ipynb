```
from __future__ import absolute_import, division, print_function, unicode_literals
try:
  # Use the %tensorflow_version magic if in colab.
  %tensorflow_version 2.x
except Exception:
  pass

import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import matplotlib.pyplot as plt
import numpy as np
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
_URL = 'http://techlinegroup.com/skinline/SKINLINE_DATASET.zip'
zip_dir = tf.keras.utils.get_file('SKINLINE_DATASET.zip', origin=_URL, extract=True)
zip_dir_base = os.path.dirname(zip_dir)
!find $zip_dir_base -type d -print
base_dir = os.path.join(os.path.dirname(zip_dir), 'SKINLINE_DATASET')
train_dir = os.path.join(base_dir, 'Train')
validation_dir = os.path.join(base_dir, 'Validate')

train_Acne_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Acne') 
train_Boil_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Boil') 
train_Eczema_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Eczema') 
train_Keratosis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Keratosis') 
train_Nail_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Nail') 
train_Rosacea_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Rosacea') 
train_Viral_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Viral') 
train_Bacteria_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Bacteria') 
train_Bullous_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Bullous') 
train_Fungal_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Fungal') 
train_Lesion_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Lesion') 
train_Psoriasis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Psoriasis') 
train_Scabies_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Scabies') 
train_Bites_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Bites') 
train_Dermatitis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Dermatitis') 
train_Herpes_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Herpes') 
train_Melanoma_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Melanoma') 
train_Rash_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Rash') 
train_Vasculitis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Train/Vasculitis') 

validation_Acne_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Acne') 
validation_Boil_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Boil') 
validation_Eczema_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Eczema') 
validation_Keratosis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Keratosis') 
validation_Nail_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Nail') 
validation_Rosacea_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Rosacea') 
validation_Viral_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Viral') 
validation_Bacteria_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Bacteria') 
validation_Bullous_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Bullous') 
validation_Fungal_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Fungal') 
validation_Lesion_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Lesion') 
validation_Psoriasis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Psoriasis') 
validation_Scabies_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Scabies') 
validation_Bites_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Bites') 
validation_Dermatitis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Dermatitis') 
validation_Herpes_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Herpes') 
validation_Melanoma_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Melanoma') 
validation_Rash_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Rash') 
validation_Vasculitis_dir = os.path.join(train_dir, '/root/.keras/datasets/SKINLINE_DATASET/Validate/Vasculitis') 

num_Acne_tr = len(os.listdir(train_Acne_dir)) 
num_Boil_tr = len(os.listdir(train_Boil_dir)) 
num_Eczema_tr = len(os.listdir(train_Eczema_dir)) 
num_Keratosis_tr = len(os.listdir(train_Keratosis_dir)) 
num_Nail_tr = len(os.listdir(train_Nail_dir)) 
num_Rosacea_tr = len(os.listdir(train_Rosacea_dir)) 
num_Viral_tr = len(os.listdir(train_Viral_dir)) 
num_Bacteria_tr = len(os.listdir(train_Bacteria_dir)) 
num_Bullous_tr = len(os.listdir(train_Bullous_dir)) 
num_Fungal_tr = len(os.listdir(train_Fungal_dir)) 
num_Lesion_tr = len(os.listdir(train_Lesion_dir)) 
num_Psoriasis_tr = len(os.listdir(train_Psoriasis_dir)) 
num_Scabies_tr = len(os.listdir(train_Scabies_dir)) 
num_Bites_tr = len(os.listdir(train_Bites_dir)) 
num_Dermatitis_tr = len(os.listdir(train_Dermatitis_dir)) 
num_Herpes_tr = len(os.listdir(train_Herpes_dir)) 
num_Melanoma_tr = len(os.listdir(train_Melanoma_dir)) 
num_Rash_tr = len(os.listdir(train_Rash_dir)) 
num_Vasculitis_tr = len(os.listdir(train_Vasculitis_dir)) 

num_Acne_val = len(os.listdir(validation_Acne_dir)) 
num_Boil_val = len(os.listdir(validation_Boil_dir)) 
num_Eczema_val = len(os.listdir(validation_Eczema_dir)) 
num_Keratosis_val = len(os.listdir(validation_Keratosis_dir)) 
num_Nail_val = len(os.listdir(validation_Nail_dir)) 
num_Rosacea_val = len(os.listdir(validation_Rosacea_dir)) 
num_Viral_val = len(os.listdir(validation_Viral_dir)) 
num_Bacteria_val = len(os.listdir(validation_Bacteria_dir)) 
num_Bullous_val = len(os.listdir(validation_Bullous_dir)) 
num_Fungal_val = len(os.listdir(validation_Fungal_dir)) 
num_Lesion_val = len(os.listdir(validation_Lesion_dir)) 
num_Psoriasis_val = len(os.listdir(validation_Psoriasis_dir)) 
num_Scabies_val = len(os.listdir(validation_Scabies_dir)) 
num_Bites_val = len(os.listdir(validation_Bites_dir)) 
num_Dermatitis_val = len(os.listdir(validation_Dermatitis_dir)) 
num_Herpes_val = len(os.listdir(validation_Herpes_dir)) 
num_Melanoma_val = len(os.listdir(validation_Melanoma_dir)) 
num_Rash_val = len(os.listdir(validation_Rash_dir)) 
num_Vasculitis_val = len(os.listdir(validation_Vasculitis_dir)) 

total_train = num_Acne_tr + num_Boil_tr +num_Eczema_tr +num_Keratosis_tr + num_Nail_tr +num_Rosacea_tr +num_Viral_tr +num_Bacteria_tr + num_Bullous_tr + num_Fungal_tr +num_Lesion_tr +num_Psoriasis_tr +num_Scabies_tr +num_Bites_tr +num_Dermatitis_tr +num_Herpes_tr +num_Melanoma_tr +num_Rash_tr +num_Vasculitis_tr 

total_val = num_Acne_val + num_Boil_val +num_Eczema_val +num_Keratosis_val + num_Nail_val +num_Rosacea_val +num_Viral_val +num_Bacteria_val + num_Bullous_val + num_Fungal_val +num_Lesion_val +num_Psoriasis_val +num_Scabies_val +num_Bites_val +num_Dermatitis_val +num_Herpes_val +num_Melanoma_val + num_Rash_val +num_Vasculitis_val 

print('total training Acne images:', num_Acne_tr)
print('total training Boil images:', num_Boil_tr)
print('total training Eczema images:', num_Eczema_tr)
print('total training Keratosis images:', num_Keratosis_tr)
print('total training Nail images:', num_Nail_tr)
print('total training Rosacea images:', num_Rosacea_tr)
print('total training Viral images:', num_Viral_tr)
print('total training Bacteria images:', num_Bacteria_tr)
print('total training Bullous images:', num_Bullous_tr)
print('total training Fungal images:', num_Fungal_tr)
print('total training Lesion images:', num_Lesion_tr)
print('total training Psoriasis images:', num_Psoriasis_tr)
print('total training Scabies images:', num_Scabies_tr)
print('total training Bites images:', num_Bites_tr)
print('total training Dermatitis images:', num_Dermatitis_tr)
print('total training Herpes images:', num_Herpes_tr)
print('total training Melanoma images:', num_Melanoma_tr)
print('total training Rash images:', num_Rash_tr)
print('total training Vasculitis images:', num_Vasculitis_tr)

print('total validation Acne images:', num_Acne_val)
print('total validation Boil images:', num_Boil_val)
print('total validation Eczema images:', num_Eczema_val)
print('total validation Keratosis images:', num_Keratosis_val)
print('total validation Nail images:', num_Nail_val)
print('total validation Rosacea images:', num_Rosacea_val)
print('total validation Viral images:', num_Viral_val)
print('total validation Bacteria images:', num_Bacteria_val)
print('total validation Bullous images:', num_Bullous_val)
print('total validation Fungal images:', num_Fungal_val)
print('total validation Lesion images:', num_Lesion_val)
print('total validation Psoriasis images:', num_Psoriasis_val)
print('total validation Scabies images:', num_Scabies_val)
print('total validation Bites images:', num_Bites_val)
print('total validation Dermatitis images:', num_Dermatitis_val)
print('total validation Herpes images:', num_Herpes_val)
print('total validation Melanoma images:', num_Melanoma_val)
print('total validation Rash images:', num_Rash_val)
print('total validation Vasculitis images:', num_Vasculitis_val)

print("--")
print("Total training images:", total_train)
print("Total validation images:", total_val)

BATCH_SIZE = 1  # Number of training examples to process before updating our models variables
IMG_SHAPE  = 150  # Our training data consists of images with width of 150 pixels and height of 150 pixels

train_image_generator      = ImageDataGenerator(rescale=1./255)  # Generator for our training data
validation_image_generator = ImageDataGenerator(rescale=1./255)  # Generator for our validation data

train_data_gen = train_image_generator.flow_from_directory(batch_size=BATCH_SIZE,
                                                           directory=train_dir,
                                                           shuffle=True,
                                                           target_size=(IMG_SHAPE,IMG_SHAPE), #(150,150)
                                                           class_mode="sparse" )

val_data_gen = validation_image_generator.flow_from_directory(batch_size=BATCH_SIZE,
                                                           directory=validation_dir,
                                                           shuffle=False,
                                                           target_size=(IMG_SHAPE,IMG_SHAPE),
                                                           class_mode='sparse')


#preparing the model
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
      
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),

    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(19, activation='softmax') # no of probabilities
])
    
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
    
model.summary()

#training the model
EPOCHS = 5
history = model.fit_generator(
    train_data_gen,
    steps_per_epoch=int(np.ceil(total_train / float(BATCH_SIZE))),
    epochs=EPOCHS,
    validation_data=val_data_gen,
    validation_steps=int(np.ceil(total_val / float(BATCH_SIZE)))
)

#visualizing result
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.savefig('./foo.png')
plt.show()

```
